# Objective Function

```python
def yangs4(x):
    # f = (a-b).c
    a = np.sum(np.square(np.sin(x)))
    b = np.exp(-np.sum(np.square(x)))
    c = np.exp(-np.sum(np.square(np.sin(np.sqrt(np.abs(x))))))
    return (a-b)*c
```

```python
n = 5
x0 = n * [0.1]
sigma0 = 0.1
# evolution strategy
es = CMAEvolutionStrategy(x0, sigma0)
optSol = es.optimize(yangs4)
```

```python
optSol.result_pretty()
```
